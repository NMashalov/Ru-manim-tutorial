```python
%%manim -pqh -v WARNING PhysicsProblem

from manim import *

class PhysicsProblem(Scene):
    def construct(self):
        #  1:  
        problem_statement = Paragraph(
            " 9.1.3",
            "      .",
            "        ?",
            "      ?",
            alignment="center",
            font_size=24
        ).scale(0.8)
        self.play(Write(problem_statement))
        self.wait(4)
        self.play(FadeOut(problem_statement))

        #  2:    
        title = Text("   :", font_size=30)
        self.play(Write(title))
        self.wait(1)

        eq_motion = MathTex("y(t)=B \sin (\omega t)")
        self.play(ReplacementTransform(title, eq_motion))
        self.wait(1)

        self.play(eq_motion.animate.to_edge(UP + LEFT))

        #   Oy    
        y_axis = NumberLine(x_range=[-10, 10], length=5, color=WHITE, include_numbers=False, include_ticks=True).rotate(PI / 2)
        y_axis.add_tip().shift(UP * 0.2)
        y_axis_label = MathTex("y", color=WHITE).next_to(y_axis.get_end(),LEFT)
        self.play(Create(y_axis), Create(y_axis_label))

        #      
        platform = Rectangle(width=2, height=0.2, fill_color=BLUE, fill_opacity=0.5).move_to(y_axis.get_center())
        self.play(Create(platform))

        #    
        def oscillate(t, amplitude=2, frequency=1):
            return amplitude * np.sin(frequency * t)

        #   
        time_0 = self.renderer.time
        self.play(
            UpdateFromFunc(
                platform,
                lambda obj: obj.move_to(
                    np.array([0, oscillate(self.renderer.time - time_0, amplitude=2, frequency=1), 0])
                ),
            ),
            run_time=2*np.pi,
            rate_func=linear,
        )
        self.wait(2)

        #   Oy         
        y_axis_target = y_axis.copy().scale(0.5).to_corner(UP + RIGHT)
        platform_target = platform.copy().scale(0.5).move_to(y_axis_target.get_center())
        y_axis_label_target = y_axis_label.copy().scale(0.5).next_to(y_axis_target.get_end(), LEFT)
        self.play(
            Transform(y_axis, y_axis_target),
            Transform(platform, platform_target),
            Transform(y_axis_label, y_axis_label_target),
        )
        self.wait(2)

        #  
        eq_acceleration = MathTex(r"a_y(t)=\frac{d^2 y}{dt^2}=-\omega^2 B \sin (\omega t)").scale(1.2)
        self.play(Write(eq_acceleration))
        self.wait(2)
        eq_acceleration_target = eq_acceleration.copy().scale(0.6).next_to(eq_motion, DOWN)
        self.play(Transform(eq_acceleration, eq_acceleration_target))
        self.wait(2)

        #  3:   
        platform_center = Rectangle(width=2, height=0.2, fill_color=BLUE, fill_opacity=0.5)
        coin = Circle(radius=0.1, color=YELLOW, fill_opacity=0.7).move_to(platform_center.get_top() + UP * 0.1)
        self.play(Create(platform_center), Create(coin))
        self.wait(1)

        #  
        gravity = Arrow(start=coin.get_bottom(), end=coin.get_bottom() + DOWN * 0.5, buff=0, color=RED)
        normal_force = Arrow(start=coin.get_top(), end=coin.get_top() + UP * 0.5, buff=0, color=GREEN)
        gravity_label = MathTex(r"m \vec{g}").next_to(gravity, DOWN)
        normal_force_label = MathTex(r"\vec{N}").next_to(normal_force, UP)

        self.play(Create(gravity), Create(normal_force), Write(gravity_label), Write(normal_force_label))
        self.wait(2)

        #  
        release_condition = VGroup(
            Text(" :", font_size=30),
            MathTex("N=0")
        ).arrange(DOWN).to_edge(LEFT)
        self.play(Write(release_condition))
        self.wait(2)

        # 2-  
        newton_law = VGroup(
            Text("2-  :", font_size=30),
            MathTex(r"m\vec{g}+\vec{N}=m\vec{a}"),
            MathTex(r"-mg=ma_y \Rightarrow a_y=-g")
        ).arrange(DOWN).to_edge(LEFT).shift(DOWN*2)
        self.play(Write(newton_law))
        self.wait(3)

        self.play(*[FadeOut(mob) for mob in self.mobjects])  #      

        #  4:   ,   
        usl = Text("  ,   :", font_size=30)
        eq1 = MathTex(r"a_y(t)=-g")
        eq2 = MathTex(r"-\omega^2 B \sin(\omega t) = -g")
        eq3 = MathTex(r"B_{\text{min}} = \frac{g}{\omega^2}")
        eq4 = MathTex(r"t_1 = \frac{1}{\omega}\arcsin\left(\frac{g}{\omega^2 B}\right)")

        equations = VGroup(usl, eq1, eq2).arrange(DOWN).to_edge(UP)
        self.play(Write(equations))
        self.wait(1)

        arrow1 = Arrow(eq2.get_bottom() + DOWN * 0.5, eq2.get_bottom() + DOWN * 2 + LEFT * 2, buff=0.1)
        arrow2 = Arrow(eq2.get_bottom() + DOWN * 0.5, eq2.get_bottom() + DOWN * 2 + RIGHT * 2, buff=0.1)
        self.play(GrowArrow(arrow1), GrowArrow(arrow2))
        self.wait(1)

        eq3_box = SurroundingRectangle(eq3, color=BLUE)
        eq3_group = VGroup(eq3, eq3_box)
        eq3_group.next_to(arrow1.get_end(), DOWN, buff=0.5)
        self.play(Write(eq3), Create(eq3_box))
        self.wait(1)

        eq4_box = SurroundingRectangle(eq4, color=YELLOW)
        eq4_group = VGroup(eq4, eq4_box)
        eq4_group.next_to(arrow2.get_end(), DOWN, buff=0.5)
        self.play(Write(eq4), Create(eq4_box))
        self.wait(1)

        #    ,  eq4  .
        other_mobs = Group(*[mob for mob in self.mobjects if mob != eq4])

        self.play(
            FadeOut(other_mobs)
        )

        #  eq4  .
        self.play(
            eq4.animate.to_edge(UP).set_x(0)
        )
        self.wait(1)

        #    
        axes = Axes(
          x_range=[0, 2 * PI, PI / 2],
          y_range=[-10, 10, 2],
            axis_config={"color": WHITE},
          x_length=10,
          y_length=6
        ).shift(DOWN * 0.8)

        labels = axes.get_axis_labels(x_label="t", y_label="a_y(t)")
        graph = axes.plot(lambda t: -10 * np.sin(t), color=BLUE)
        horizontal_line = axes.plot(lambda t: -7, color=RED, x_range=[0, 2 * PI])
        intersection_point = axes.c2p(np.arcsin(7 / 10), -7)
        intersection_dot = Dot(intersection_point, color=YELLOW)
        vertical_line = DashedLine(intersection_point, axes.c2p(np.arcsin(7 / 10), 0), color=YELLOW)
        t1_point = Dot(axes.c2p(np.arcsin(7 / 10), 0), color=YELLOW)
        t1_label = MathTex("t_1").next_to(t1_point, DOWN)

        self.play(Create(axes), Write(labels))
        self.play(Create(graph), Create(horizontal_line))
        self.play(Create(intersection_dot), Create(vertical_line), Create(t1_point), Write(t1_label))
        self.wait(3)

        self.play(*[FadeOut(mob) for mob in self.mobjects])  #      

        #  5:         
        #  
        title5 = Paragraph(
            "    ",
            "   ",
            alignment="center",
            font_size=24
        ).scale(0.8).to_edge(UP)
        self.play(Write(title5))
        self.wait(2)

        #  
        eq5_1 = MathTex(r"y_0 = B \sin(\omega t_1) = \frac{g}{\omega^2}")
        eq5_2 = MathTex(r"v_{0y} = \omega B \cos(\omega t_1) = \omega B \sqrt{1-\frac{g^2}{(\omega^2 B)^2}} = \sqrt{\omega^2 B^2 - \frac{g^2}{\omega^2}}")

        equations5 = VGroup(eq5_1, eq5_2).arrange(DOWN)
        self.play(Write(equations5))
        self.wait(2)

        #  
        eq_transformed = MathTex(r"y(t>t_1)=y_0 + v_{0y} (t-t_1) - \frac{g (t-t_1)^2}{2}")
        self.play(
            Transform(equations5, eq_transformed.next_to(title5, DOWN, buff=0.5))
        )
        self.wait(2)

        #   
        axes = Axes(
            x_range=[0, 2 * PI, PI / 2],
            y_range=[-15, 15, 2],  #     y
            x_length=10,
            y_length=6,  #    y
            tips=True,
        ).shift(DOWN)
        axes_labels = axes.get_axis_labels(x_label="t", y_label="y(t)")
        self.play(
            Write(axes),
            Write(axes_labels)
        )
        self.wait(1)

        # 
        t1 = np.arcsin(7 / 10)  #  
        y1 = lambda t: 10 * np.sin(t)
        y2 = lambda t: 7 + np.sqrt(100 - 49) * (t - t1) - 7 / 2 * (t - t1) ** 2

        graph1 = axes.plot(y1, [0, t1], color=BLUE)
        self.play(Create(graph1))
        self.wait(1)

        graph2 = axes.plot(y2, [t1, 4.1], color=YELLOW)
        graph1_dashed = axes.plot(y1, [t1, 4.1], color=BLUE_B)
        self.play(Create(graph2), Create(graph1_dashed))
        self.wait(2)

        #     
        self.play(*[FadeOut(mob) for mob in self.mobjects])



        #  6:   
        #  
        start_text = Text('   :', font_size=30).to_edge(UP)
        self.play(Write(start_text))

        #  
        eq6_1 = MathTex(r"\frac{mv_{0y}^2}{2} + mgy_0 = mgy_{\text{max}}").next_to(start_text,DOWN)
        self.play(Write(eq6_1))

        # 
        arrow = Arrow(eq6_1.get_bottom(), DOWN, color=WHITE)
        self.play(GrowArrow(arrow))

        #  
        eq6_2 = MathTex(r"y_{\text{max}} = \frac{\omega^2 B^2 + g^2}{2g \omega^2}").next_to(arrow.get_end(), DOWN)
        framebox = SurroundingRectangle(eq6_2, color=GREEN)
        self.play(Write(eq6_2), FadeIn(framebox))

        self.wait(2)

        #   
        self.play(*[FadeOut(mob) for mob in self.mobjects])

        #  
        text = Text(':', font_size=30).to_edge(UP)
        answer = MathTex(r"B_{\text{min}} = \frac{g}{\omega^2}", r",", r"y_{\text{max}} = \frac{\omega^2 B^2 + g^2}{2g \omega^2}").scale(1.2)
        self.play(
            Write(text),
            Write(answer)
            )

        self.wait(2)
```

```python
%%manim -qm -v WARNING OscillatingPlatform

class OscillatingPlatform(Scene):
    def construct(self):
        title = Text("   :", font_size=30)
        self.play(Write(title))
        self.wait(1)

        eq_motion = MathTex("y(t)=B \sin (\omega t)")
        self.play(ReplacementTransform(title, eq_motion))
        self.wait(1)

        self.play(eq_motion.animate.to_edge(UP + LEFT))

        #   Oy    
        y_axis = NumberLine(x_range=[-10, 10], length=5, color=WHITE, include_numbers=False, include_ticks=True).rotate(PI / 2)
        y_axis.add_tip().shift(UP * 0.2)
        y_axis_label = MathTex("y", color=WHITE).next_to(y_axis.get_end(),LEFT)
        self.play(Create(y_axis), Create(y_axis_label))

        #      
        platform = Rectangle(width=2, height=0.2, fill_color=BLUE, fill_opacity=0.5).move_to(y_axis.get_center())
        self.play(Create(platform))

        #    
        def oscillate(t, amplitude=2, frequency=1):
            return amplitude * np.sin(frequency * t)

        #   
        time_0 = self.renderer.time
        self.play(
            UpdateFromFunc(
                platform,
                lambda obj: obj.move_to(
                    np.array([0, oscillate(self.renderer.time - time_0, amplitude=2, frequency=1), 0])
                ),
            ),
            run_time=2*np.pi,
            rate_func=linear,
        )
        self.wait(2)

        #   Oy         
        y_axis_target = y_axis.copy().scale(0.5).to_corner(UP + RIGHT)
        platform_target = platform.copy().scale(0.5).move_to(y_axis_target.get_center())
        y_axis_label_target = y_axis_label.copy().scale(0.5).next_to(y_axis_target.get_end(), LEFT)
        self.play(
            Transform(y_axis, y_axis_target),
            Transform(platform, platform_target),
            Transform(y_axis_label, y_axis_label_target),
        )
        self.wait(2)

        #  
        eq_acceleration = MathTex(r"a_y(t)=\frac{d^2 y}{dt^2}=-\omega^2 B \sin (\omega t)").scale(1.2)
        self.play(Write(eq_acceleration))
        self.wait(2)
        eq_acceleration_target = eq_acceleration.copy().scale(0.6).next_to(eq_motion, DOWN)
        self.play(Transform(eq_acceleration, eq_acceleration_target))
        self.wait(2)
```
