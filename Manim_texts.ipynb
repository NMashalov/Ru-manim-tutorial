### Трансофрмации

```python
%%manim -qm  -v WARNING OpManim
class OpManim(Scene):
    def construct(self):

        text = Text(r"Transform позволяет гладко переходить")

        self.play(
            Write(text),
        )
        self.wait(duration=2)

        new_text = Text(r"В новый текст")

        self.play(
            Transform(text, new_text)
        )
        self.wait(duration=2)
```

```python
%%manim -qm  -v WARNING OpManim
class OpManim(Scene):
    def construct(self):

        text = MathTex(r"\int_\sigma d\omega = \int_{\partial \sigma} \omega").to_corner(UP)

        instruction = Text('Вверх .to_corner(UP)')

        self.play(
            Write(text),
            Write(instruction)
        )
        self.wait(duration=2)

        new_text = MathTex(r"\int_{\Sigma} \text{rot} \mathbf{F} d \mathbf{\Sigma} = \int_{\partial \Sigma} F \cdot d \mathbf{r}").to_corner(DOWN+RIGHT)

        self.play(
            Transform(instruction, Text('Вправо вниз .to_corner(DOWN+RIGHT)')),
            Transform(text, new_text)
        )
        self.wait(duration=2)
```

## Домашнее задание

Запишите доказательство небольшой теоремы используя текст

Постарайтесь использовать анимацию для

В этом примере я использовал переход, чтобы запутать зрителя в доказательстве "4=5"

```python
%%manim -qm -v WARNING Obfuscation

'''
Анимация будет состоять из двух сцен
Мы постараемся запутать зрителя
'''



first_part = [
    r'\sqrt{-20}=\sqrt{-20}',
    r'\sqrt{16-36} = \sqrt{25-45}',

]

diversion = r'Добавим $\frac{81}{4}$ к обоим сторонам'

second_part = [
    r'\sqrt{16-36+\frac{81}{4}} = \sqrt{25-45+\frac{81}{4}}',
    r'\sqrt{4^2 - 2 \cdot 4 \cdot \frac{9}{2} + \left(\frac{9}{2}\right)^2} = \sqrt{5^2 - 2 \cdot 5 \cdot \frac{9}{2} + \left(\frac{9}{2}\right)^2}',
    r'\sqrt{(4-\frac{9}{2})^2} = \sqrt{(5-\frac{9}{2})^2}',
    r'4-\frac{9}{2}=5-\frac{9}{2}',
    r'4=5'
]


class Obfuscation(Scene):
    def construct(self):

        title = Tex(r"Доказательство $4=5$")
        # arrange(DOWN) задает шаг между
        base = Group(*map(MathTex,first_part)).arrange(DOWN)
        Group(title,base).arrange(DOWN)

        self.play(
            Write(title),
            LaggedStart(FadeIn(base, shift=DOWN)),
        )
        self.wait(duration=2)

        transform_title = Tex(diversion)
        new_base = VGroup(*map(MathTex,second_part)).arrange(DOWN)

        Group(transform_title,new_base).arrange(DOWN)

        self.play(
            Transform(title, transform_title),
            LaggedStart(*(FadeOut(obj, shift=DOWN) for obj in base)),
            FadeIn(new_base),
        )
        self.wait()
```

## Анонс. Работа с гоеометрией

```python
%%manim -qm -v WARNING ContinuousMotion
from manim import UR,StreamLines
import numpy as np

class ContinuousMotion(Scene):
    def construct(self):
        func = lambda pos: np.sin(pos[0] / 2) * UR + np.cos(pos[1] / 2) * LEFT
        stream_lines = StreamLines(func, stroke_width=2, max_anchors_per_line=30)
        self.add(stream_lines)
        stream_lines.start_animation(warm_up=False, flow_speed=1.5)
        self.wait(stream_lines.virtual_time / stream_lines.flow_speed)
```
